```python
#Importación de librerías
import requests
import sqlalchemy
import pandas as pd
from bs4 import BeautifulSoup
from sqlalchemy.types import Integer, Text, Float
import csv
```

```python
#Variable con la dirección del sitio a hacer scraping
url = "https://books.toscrape.com/"
```

```python
#Realizar la petición
books_to_scrape = requests.get(url)
```

```python
#Parsear la información, mediante una instancia de beautiful soup
soup = BeautifulSoup(books_to_scrape.text, 'lxml')
#Confirmar el tipo de dato de la "sopa" creada
type(soup)
```

```python
#Imprimir con formato
print(soup.prettify())
```

```python
#Apuntar a los artículos para extraerlos
articulos = soup.find_all('article', attrs={'class':'product_pod'})
articulos
```

```python
# Lista temporal para guardar los libros
libros = []
```

```python
books_df
```

```python
#iterar los articulos y obtener las variables
for articulo in articulos:
    titulo_libro = articulo.h3.a.get('title')
    imagen_libro = articulo.img.get('src')
    precio_libro_raw = articulo.find('p', attrs={'class':'price_color'}).get_text()[1:]
    precio_libro = precio_libro_raw[1:]
    divisa = precio_libro_raw[0]
    temp_row = [titulo_libro, imagen_libro, precio_libro]
    libros.append(temp_row)
    print(temp_row)
```

```python
books_df = pd.DataFrame(libros,columns=['titulo_libro','imagen_libro','precio_libro', 'divisa'])
```

```python
# Inspeccionar el dataframe
books_df
```

```python
# Generar la conexión 
url = 'postgresql+psycopg2://postgres:TaekoBB@127.0.0.1:5432/books_to_scrape'
engine = sqlalchemy.create_engine(url)
```

```python
# Transformar un dataframe a una base de datos
books_df.to_sql(
    "libros", 
    con = engine,
    if_exists = "replace",
    schema='public',   
    chunksize=1000,
    dtype={
       "id": Integer,
       "titulo_libro": Text,
       "imagen_libro": Text,
       "precio_libro": Text,
       "divisa": Float
     }
)
```
