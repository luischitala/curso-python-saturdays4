```python
# Importación de librerías
import numpy as np
import pandas as pd
# Graficación
import matplotlib.pyplot as plt
import seaborn as sns
# ML Feature selection
from sklearn.model_selection import train_test_split
# Modelo de regresión logística
from sklearn.linear_model import LogisticRegression
# Transformación de datos
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
# Evaluación del modelo
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Modelo de regresión logística

Permite modelar el comportamiento de una variable dependiente y las variables independientes.

Es un algoritmo de aprendizaje supervisado de clasificación.

Útil para predicción de variables objetivo:
- Binarias

## Carga del dataset

```python
# Carga de dataset
df = pd.read_csv('Iris.csv')

# Verificación de valores
df.head()
```

### Descubrimiento de los datos

```python
# Obtener la información de las características e integridad de los datos
df.info()
```

```python
# Estadísticos básicos
df.describe()
```

```python
# Verificación de valores vacíos
print(df.isna().sum())
```

```python
# Eliminación del id, ya que no será considerado por el modelo
df.drop(columns = ['Id'], axis = 1, inplace = True)
df
```

```python
# Conteo de especies
print("Las especies únicas en el dataset son:", df['Species'].unique())
```

### Visualización de los datos

```python
plt.figure(figsize = (10, 7))
sns.scatterplot(x = 'SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', data = df, s = 65)
plt.legend(bbox_to_anchor = (1, 1), loc = 2, fontsize = 15)
plt.title('Longitud del Sépalo VS Ancho del Sépalo', fontsize = 20)
plt.xlabel('Longitud del Sépalo', fontsize = 15)
plt.ylabel('Ancho del Sépalo', fontsize = 15)
plt.show()
```

### Correlación de los datos

```python
plt.figure(figsize = (15, 10))
sns.heatmap(df.corr(), cmap = 'Blues', square = True, annot = True)
plt.title("Correlación de los valores", size = 20)
plt.show()
```

### Manejo y escalado de los datos

```python
# Separación de los datos en conjuntos de características y característica a predecir
X = df.drop(['Species'], axis = 1)
y = df['Species']
print(X)
print("--"*35)
print(y)
```

#### Escalado de los datos

El escalado de datos es importante debido a los siguientes puntos:  
- Requisito del modelo
- Mejoramiento de la precisión
- Ahorro de procesamiento

```python
"""
Ejemplo de escalado estandar de datos
Su objetivo es procesar los datos para lograr estandarizar los datos 
Que precisamente busca lograr una media y desviación estandar 
en el conjunto de datos final

# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
"""

X1 = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X1) 
# Datos iniciales
print(X1)
# Datos escalados
print(scaled_data)

# Verificación de la media
print(f'La media de los datos transformados es {scaled_data.mean(axis = 0)}')

# Verificación de la desviación estándar
print(f'La desviación estándar de los datos transformados es {scaled_data.std(axis = 0)}')
```

```python
# Inicializamos un objeto de la clase StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Forma transformada de las características del dataset
print(X)
```

```python
# Verificación de la media
print(f'La media de los datos transformados es {X.mean(axis = 0)}')

# Verificación de la desviación estándar
print(f'La desviación estándar de los datos transformados es {X.std(axis = 0)}')
```

#### Codificar las especies de flores

```python
# Generar una instancia del objeto labelEncoder
# Para transformar los 3 valores de las especies a números: 0,1,2
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

#### Separación de los conjuntos de datos de entrenamiento y prueba

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 0)
```

### Modelo de regresión logística

```python
# Instanciación del objeto 
lr = LogisticRegression()
# Ingreso de datos de entrenamiento
lr.fit(X_train, y_train)
# Predicción de valores usando el método predict y los datos de prueba
y_pred = lr.predict(X_test)
# Obtención de la precisión
accuracy = lr.score(X_test, y_test)
print("La precisión del modelo es:", accuracy * 100, "%")
```

### Evaluación del modelo

1. Precisión: Con la métrica de precisión podemos medir la calidad del modelo de machine learning en tareas de clasificación.

2. Recall(Exhaustividad): La métrica de exhaustividad nos va a informar sobre la cantidad que el modelo de machine learning es capaz de identificar. 

3. F1 Score: El valor F1 se utiliza para combinar las medidas de precision y recall en un sólo valor. Esto es práctico porque hace más fácil el poder comparar el rendimiento combinado de la precisión y la exhaustividad entre varias soluciones. 

4. Precisión (Accuracy): La exactitud (accuracy) mide el porcentaje de casos que el modelo ha acertado. Esta es una de las métricas más usadas y favoritas 

5. Support: El número de muestras de respuesta correcta dentro de cada clase. 

Referencia: https://www.iartificial.net/precision-recall-f1-accuracy-en-clasificacion/

```python
cl_rep = classification_report(y_test, y_pred)
print("\n El reporte de clasificación de la regresión logística es:", cl_rep, sep = "\n")
cm_model = confusion_matrix(y_test, y_pred)
plt.figure(figsize = (8, 6))
sns.heatmap(cm_model, annot = True, cmap = 'Blues', annot_kws = {'size': 15}, square = True)
plt.title('Matriz de confusión del modelo de regresión logística ', size = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()    
```
